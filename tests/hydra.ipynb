```
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (7,7)
```

```
subcircuit_16_7_size = [2**i for i in range(9)]
protocol_16_7_time = [405.5, 197.4, 99.4, 49.8, 25.5, 14.9, 12.1, 13.0, 15.8] #97.02%
```

```
plt.title('Subcircuit Protocol Evaluation (2^16 Depth, 2^7 Width)')

plt.xscale('log', basex=2)
plt.xlabel('Number of Subcircuits')

plt.ylabel('Protocol Time (s)')

plt.plot(subcircuit_16_7_size,protocol_16_7_time,'-o')
plt.grid(True)
plt.show()
```
